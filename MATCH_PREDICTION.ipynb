**MODEL**

```python
X_dataframe = pd.read_csv('MATCH_PREDICTION_X.csv', header = None)
Y_dataframe = pd.read_csv('MATCH_PREDICTION_Y.csv', header = None)

X = X_dataframe.values
Y = Y_dataframe.values
min_max_scaler = preprocessing.MinMaxScaler()   
X_scale = min_max_scaler.fit_transform(X)      
encoded_y = np_utils.to_categorical(Y)         
X_training, X_testing, Y_training, Y_testing = train_test_split(X_scale, encoded_y, test_size=0.20, random_state=10)
print(len(X_training))
print(len(X_testing))
```

```python
train_scores = np.zeros(10)
test_scores = np.zeros(10)

for i in range(10):

  from keras.regularizers import L1, L2
  from keras.layers import Dropout 

  # Creating the model
  model = Sequential()
  model.add(Dense(2048, input_shape=(8,), activation = 'relu'      parameters : 2048 * 8 + 2048                   
                  ,kernel_regularizer=L1(0.00001)    cost = error cost + (alpha) * regularization
                  ))
  model.add(Dense(1024, activation = 'relu'                           parameters : 1024 * 2048 + bias = 1024   
                  ,kernel_regularizer=L1(0.0001)
                  ))
  # model.add(Dropout(rate = 0.1))
  model.add(Dense(512, activation = 'relu'
                  ,kernel_regularizer=L2(0.0001)
                  ))
  # model.add(Dropout(rate = 0.1))
  model.add(Dense(256, activation = 'relu' 
                  ,kernel_regularizer=L2(0.0001)
                  ))
  # model.add(Dropout(rate = 0.1))
  model.add(Dense(128, activation = 'relu'
                  # ,kernel_regularizer=L2(0.001)
  ))
  # # model.add(Dropout(rate = 0.1))
  model.add(Dense(64, activation = 'relu'
                  # ,kernel_regularizer=L1(0.0002)
  ))
  model.add(Dense(32, activation = 'relu'
                  # ,kernel_regularizer=L1(0.0002)
  ))
  model.add(Dense(16, activation = 'relu'
                  # ,kernel_regularizer=L1(0.0002)
  ))
  model.add(Dense(8, activation = 'relu'
                  # ,kernel_regularizer=L1(0.0002)
  ))
  model.add(Dense(4, activation = 'relu'
                  # ,kernel_regularizer=L1(0.0002)
  ))
  model.add(Dense(2, activation='softmax'                     
                  # ,kernel_regularizer=L1(0.0002)
  ))

  # model.summary()   #gives a summary of the model

  opt=optimizers.Adam(learning_rate=0.00001)    
  model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])         
  number_of_epochs = 400
  batch_size_val = 256
  hist = model.fit(X_training, Y_training,  epochs=number_of_epochs, batch_size = batch_size_val, verbose = 2
                   ,validation_data = (X_testing,Y_testing)
                   )
  train_scores[i] = model.evaluate(X_training, Y_training)[1]
  test_scores[i] = model.evaluate(X_testing, Y_testing)[1]
```

```python
print("Average Training Accuracy : ",np.average(train_scores))
print("Average Testing Accuracy : ",np.average(test_scores))
```

```python
class_names = ["First", "Second"]
train_prob = model.predict(X_training)
train_pred = np.argmax(train_prob, axis=1)
train_real = np.argmax(Y_training, axis = 1)
train_accuracy = np.mean(train_pred == train_real)
# print(train_accuracy)

# conf_mat = confusion_matrix(train_real, train_pred)
# print(pd.DataFrame(conf_mat))

# fig, ax = plot_confusion_matrix(conf_mat=conf_mat,
#                                 show_normed=True,
#                                 show_absolute=False,
#                                 class_names = class_names,
#                                 figsize=(8, 8))
# fig.show()
print("Classification Report for Training Data : ")
print(classification_report(train_real, train_pred))

test_prob = model.predict(X_testing)
test_pred = np.argmax(test_prob, axis=1)
test_real = np.argmax(Y_testing, axis = 1)
test_accuracy = np.mean(test_pred == test_real)
# print(test_accuracy)

# conf_mat = confusion_matrix(test_real, test_pred)
# print(pd.DataFrame(conf_mat))

# fig, ax = plot_confusion_matrix(conf_mat=conf_mat,
#                                 show_normed=True,
#                                 show_absolute=False,
#                                 class_names = class_names,
#                                 figsize=(8, 8))
# fig.show()
print("Classification Report for Testing Data : ")
print(classification_report(test_real, test_pred))
```
